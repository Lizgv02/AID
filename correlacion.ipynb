```python
import pandas as pd 
import seaborn as sns

import warnings

#set warnings to ignore

warnings.filterwarnings('ignore')

import os
```

```python
df = pd.read_csv('housing.csv', header=0, sep=',')
df.head(4)
```

```python
corr= df[['housing_median_age','total_rooms','total_bedrooms','population','households']].corr()

sns.heatmap(corr, cmap='Blues_r', annot=True)
```

```python
corr= df[ [ 'longitude', 'latitude'] ].corr(method='spearman')

sns.heatmap(corr, cmap='Blues_r', annot=True)
```
