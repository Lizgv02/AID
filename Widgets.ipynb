```python
import ipywidgets as widgets
from IPython.display import display, HTML
```

```python
# Estilo para hacer las cajas de texto más grandes
estilo = "<style>.widget-text { width: 400px !important; }</style>"
display(HTML(estilo))

# Crear dos cajas de texto grandes para capturar datos en la parte superior
textbox1 = widgets.Textarea(description='Dato 1:', layout=widgets.Layout(width='80%', height='100px'))
textbox2 = widgets.Textarea(description='Dato 2:', layout=widgets.Layout(width='80%', height='100px'))

# Crear dos cajas de texto grandes para capturar datos en la parte inferior
textbox3 = widgets.Textarea(description='Dato 3:', layout=widgets.Layout(width='80%', height='100px'))
textbox4 = widgets.Textarea(description='Dato 4:', layout=widgets.Layout(width='80%', height='100px'))

# Crear un botón para guardar los datos capturados
button = widgets.Button(description='Guardar')

# Función que se ejecutará al hacer clic en el botón
def guardar_datos(b):
    datos = {
        'Dato 1': textbox1.value,
        'Dato 2': textbox2.value,
        'Dato 3': textbox3.value,
        'Dato 4': textbox4.value
    }
    # Aquí se guardaran los datos ya sea en un archivo, base de datos, imprimirlos, etc.
    print("Datos guardados:", datos)

# Asignar la función al evento de clic del botón
button.on_click(guardar_datos)

# Mostrar los widgets organizados en dos filas
display(widgets.HBox([textbox1, textbox2]))
display(widgets.HBox([textbox3, textbox4]))
display(button)
```
