```python
import numpy as np
import pandas as pd
import os
import missingno as msno
import seaborn as sns 

#from ydata_profiling import ProfileReport
from matplotlib import cbook
```

```python
df = pd.read_csv('employee_salary.csv', header=0, sep=',')
```

```python
#1.-cual es el salario minimo, maximo por Gender
print('min',df.groupby(df['Gender'])['Base Salary'].min())

print('max',df.groupby(df['Gender'])['Base Salary'].max())
```

```python
#2.-cual es el salario promedio por Gender
df.groupby(df['Gender'])['Base Salary'].mean()
```

```python
#3.-cual es el salario minimo, maximo por división
print('min',df.groupby(df['Division'])['Base Salary'].min())

print('max',df.groupby(df['Division'])['Base Salary'].max())
```

```python
#4.-cual es el salario media y mediana por división
df.groupby(df['Division'])['Base Salary'].agg(['mean', 'median'])
```

```python
#5.-cual es el salario media y mediana por Deparment Name
df.groupby(df['Department Name'])['Base Salary'].agg(['mean', 'median'])
```

```python
#6.-cual es el salario sd y var por Deparment Name
df.groupby(df['Department Name'])['Base Salary'].agg(['std', 'var'])
```

```python
#7.-cual es el salario media y mediana por Deparment Name y Gender
df.groupby(['Department Name', 'Gender'])['Base Salary'].agg(['mean', 'median'])
```

```python
#8.-cual es el salario sd y var por Deparment Name y Gender
df.groupby(['Department Name', 'Gender'])['Base Salary'].agg(['std', 'var'])
```

```python
from scipy.stats import shapiro 
from scipy.stats import kstest, norm 
```

```python
sample = df['Base Salary'].sample(n=50)

print('shapiro')
stat, p_value = shapiro(sample)

print('stat', stat)
print('p-value', p_value)

if p_value > 0.05:
    print('Acepta H0: tiene una distribucion normal')
else:
    print('Acepta H1: no tiene una distribución normal')
    
print()
```

```python
print('kstest')
stat, p_value = kstest(sample, norm.cdf)

print('stat', stat)
print('p-value', p_value)

if p_value > 0.05:
    print('Acepta H0: tienen una distribucion normal')
else:
    print('Acepta H1: no tiene una distribucion normal')
    
print()
```
